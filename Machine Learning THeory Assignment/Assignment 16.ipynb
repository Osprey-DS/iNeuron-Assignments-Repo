### 2. What is the concept of simple linear regression? Give a specific example.

Ans>> The linear regression model is only applicable to the data if it has a linear relationship with the target variable. And a simple linear regression problem is where the regression model estimates the relationship with one independent variable and one dependent variable.

Formula for simple linearn regression is 

`y = b_0 + b_1*X + e`

Where `y` is the predicted value of the dependent variable or target variable.

`X` is the independent variable or predictor of the target class.

`b_0` is called intercept or bias also called it constant. 

`b_1` is called the coefficient or weights of the independent feature, as in this case we only have one independent feature that is why the coefficient is also one, it is occure as many independent variables we have in our data.

An example can be a student have as many degrees as he or she will earn more money in the future, there is a linear relationship whit degrees and earn, the more the degree is the more he will earn.
Or another example can be house size with the house price, the more house size is the more price it will be.
