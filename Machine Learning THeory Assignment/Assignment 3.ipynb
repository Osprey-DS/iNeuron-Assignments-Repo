###           2. Learning that is guided and unsupervised
Ans>> The learning with guided is called Supervised learning where we have a guide who tells the difference between objects and teach us or help us to learn how and what features a class or object has. In a supervised machine, learning data are feed features with labels or classes so it can learn the underlying pattern in order to recognize the specific object or class. That is why it is called guided learning.
And in the case of unsupervised learning or unguided learning, there are no labels, and classes are provided to the model in the time of learning, in that case, the model is try to grouping similar features in clusters or different groups in order to separate them as different classes or object so in the time of recognizing other data it can match these features and put them in appropieat cluster to define the class or label. Here there is no guide to teach the model that which feature matches the labels that, model learn by themselves by the similarity of features, that is why it is called unguided learning.


###      3. Regression and classification
Ans>> The Regression problem is all about calculating quantity or better say we are predicting continuous number like price, measurements,  in regression problem we have multiple features and we calculate weights of those feature in order to calculate the number. There are many machine learning algorithms for performing regression tasks like `Linear Regression`, `Lasso Regression`, `Randomforest Regressor` and many more even those algorithms we are doing classification task those have their regression version too.
And the Classification where we are predicting a specific class or labels from many features, where we dont find any number. There are many ways to find labels or class from the features in classification problems every algorithm have their own technique or different approaches to predict the classes. Few popular algorithms for classification tasks are `Logistic Regression`, `RandomForest`, `Support Vector machine`, and many more, these three algorithms have completely different approaches to find predict the label or they have a completely different calculation for doing the prediction. Like Logistic regression use probability to predict the labels, RandomForest has tree methods to find the class, it divided features and performing a kind of if-else process to find the class, where SupportVector use only one class to set a border between classes, although the support vector machine works well only binary classification task where we only predict only two classes.
