### 8.	What is Binomial distribution? What is the formula?

Ans>> A binomial distribution can be described as simply the probability of a True or False outcome in an experiment or survey that is repeated multiple times. The binomial is a type of distribution that has two possible outcomes. For example, a coin toss has only two possible outcomes: heads or tails, and taking a test could have two possible outcomes: pass or fail. In binomial distribution, every experiment or trial is completely independent of each other trail, which means what happens will the next trial does not depend on the current trial.

The binomial distribution formula is:

`b(x; n, P) = nCx * Px * (1 – P)n – x`

Where:

b = binomial probability

x = total number of “successes” (pass or fail, heads or tails, etc.)

P = probability of success on an individual trial

n = number of trials

### 10.	Define covariance.

Cov(X,Y) = Σ E((X-μ)E(Y-ν)) / n-1 where:

X is a random variable

E(X) = μ is the expected value (the mean) of the random variable X and

E(Y) = ν is the expected value (the mean) of the random variable Y

n = the number of items in the data set

The main problem with this measurement it can give us any range of numbers, the larger the value will be the more correlated with each other.

### 14.	What is the hypothesis? Give example.

Ans>> The hypothesis is an idea that can be tested from given information or the data, to judge a certain or given claim. The hypothesis is a core module in statistics, it helps us to ensure that a given statement how much truth and how much it is acceptable.

The hypothesis is worked on some rules and formulation, it is used one parameter called ‘p-values to measure the truthness of a certain statement.

The hypotheses statement is divide into two groups one is called

`Null Hypothesis` – this is the statement that is come to us and we have to gather proof against it.

And the second one is `Alternative Hyothesis` – this is the opposite of the null hypothesis and we try to prove it.

Note that the main aim of hypothesis testing is rejecting the ‘Null Hypothesis’ and we the experiment for it.
From the calculation, if our p-value is lower than the significance level then we can reject the Null hypotheses statement and suggest the alternative hypothesis.

And If our p-value is greater than or equal to the significance level then we fail to reject the null hypothesis statement. But that not means we accept the alternative statement, after this, we can gather more relevant data and go for further experiments.
