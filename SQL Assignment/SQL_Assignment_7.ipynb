### Q-14. Write an SQL query to clone a new table from another table.

```
                    CREATE TABLE 
                        new_table LIKE Original_table;

```
By excuting above command it will create a new or clone table like `Original_table` and for inserting the data into cloned table. we have to excute the following query...

```
                    INSERT INTO 
                        new_table 
                    SELECT 
                            * 
                    FROM Original_table
```
