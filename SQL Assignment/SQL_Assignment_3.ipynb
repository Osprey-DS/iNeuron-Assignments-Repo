### Q-1. Write an SQL query to print the FIRST_NAME from Worker table after removing white spaces from the right side.

The sql `TRIM()` remove white sapce also can remove mentioned character from the records

```
                
            SELECT 
                 RTRIM(FIRST_NAME)
            FROM 
                Worker;

```

### Q-2. Write an SQL query that fetches the unique values of DEPARTMENT from Worker table and prints its length.

Here i will use `DISTINCT` and `LENGTH()` for solving this.

```
            SELECT 
                DISTINC DEPARTMENT, LENGTH(DEPARTMENT) AS D_Length
            FROM
                Worker;
            

```

###  Q-3. Write an SQL query to fetch nth max salaries from a table.

There is various way to solve above problem, below i have shown one solution.

For getting the Nth max salary i will use `LIMIT` and `ORDER BY DESC`.


```
                        SELECT 
                            salary
                        FROM
                            salaries
                        ORDER BY salary DESC
                        LIMIT N-1,1; -- Nth highest salaries


```

here N is the number which we want the Nth salary.

e.g if we want 4th max salary then N will be 4.
