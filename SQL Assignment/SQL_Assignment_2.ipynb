### Q-3. Write an SQL query to print the name of employees having the highest salary in each department.

The answer can be varie as no database and table information is provided. I am writing this answer based on my practices database employees.

For obtaining the highest salary for each depatrment. 

```
                        SELECT 
                            d.dept_name, MAX(salary) AS max_salary
                        FROM
                            departments AS d
                                JOIN
                            dept_manager AS m ON d.dept_no = m.dept_no
                                JOIN
                            salaries AS s ON m.emp_no = s.emp_no
                        GROUP BY d.dept_name 
                        ORDER BY max_salary DESC;

                
```
