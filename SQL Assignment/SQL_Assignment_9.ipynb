### Q-17. Write an SQL query to show the top n (say 10) records of a table.

we can use `LIMIT`clause for this problem

```
SELECT 
    *
FROM employees
LIMIT 10

```

The above query will return top `10` records from the table.



### Q-18. Write an SQL query to determine the nth (say n=5) highest salary from a table.

For this problem i have used `LIMIT` and `ORDER BY` with `DESC` clause.

```
                    SELECT 
                        *
                    FROM
                        salaries
                    ORDER BY salary DESC
                    LIMIT 5;
```
