```
        SELECT
                * 
        FROM 
            table_name
        ORDER BY id DESC
        LIMIT 5) ORDER BY id ASC;
        
```
        
Here i have used `ORDER BY` for getting the last 5 recorde. i used `DESC` or Decending methode which will return the biggest number first and then smaller than that one and so on. And in last again i have used `ORDER BY` for showing the result in ascending or in normal way as it is store in database. 

Here the `id` column is primary key of the tabel and it should be unique for each record and it should be increment with each record store in the database, so the last record have the highst `id` and the first record will be lowest `id`

Here i have shown a example where i extract last 5 recods from employees tabel.
  


Notice that `emp_no` is id or `Primary key` here.
