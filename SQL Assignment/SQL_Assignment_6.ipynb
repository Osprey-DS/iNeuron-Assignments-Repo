### Q-1. Write an SQL query to print details of the Workers who have joined in Feb’2014.


```
                SELECT 
                    *
                FROM
                    Workers
                WHERE
                    YEAR(joining_date) = 2014
                        AND MONTH(joining_date) = 02;

 
```

### Q-2. Write an SQL query to fetch duplicate records having matching data in some fields of a table.



```
                        SELECT 
                            *, COUNT(*) AS dup_no
                        FROM
                            employees
                        GROUP BY first_name
                        HAVING COUNT(*) > 1;

```

Here i have used `HAVING` and `GROUP BY` caluse to detect duplicate rows..





From above image we can see that only `1275` records are returned, while the table contain more than 300k records.

The last column or `dup_no` is showing how many times one particullar record appear repeatedly.

### Q-3. How to remove duplicate rows from Employees table.


```
            WITH emp AS 
            (
            SELECT *, ROW_NUMBER() OVER(PARTITION BY first_name ORDER BY first_name) AS row_num
            FROM employees
            )
            DELETE FROM emp WHERE row_num > 1;

```

The above query will delete all matchin `first_name` except will keep one.
