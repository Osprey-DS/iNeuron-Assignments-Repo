### 3. What is a box tuple?

Ans>> Ans: Box tuple is a tuple which contains 4 integer elements, which represents a rectangular region(box) in an image.


left = The x-coordinate of the leftmost edge of the box

Top = The y-coordinate of the top edge of the box.

Right = The x-coordinate of the pixel which represents the rightmost edge of the box. It must be greater than the left value.

Bottom = The y coordinate of the pixel for the bottom most edge of the box. This integer must be greater than the top value.
